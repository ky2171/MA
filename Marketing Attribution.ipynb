# Weighting

```python
channels = ['bing','display','facebook','search','youtube']
time_list= ['2019-07','2019-08','2019-09','2019-10','2019-11','2019-12','2020-01','2020-02']
```

```python
##att_dic = att.to_dict(orient='records')
##att_dic
```

```python
result = pd.DataFrame(columns=['bing','display','facebook','search','youtube'], index = time_list)
```

```python
channel_count = [0,0,0,0,0]

for t in time_list:
    channel_count = [0,0,0,0,0]
    att_t = att[att['account_creation_date'] == t].reset_index(drop = True)
    for i in range(0, len(att_t)):
        a = att_t['attribution_technical'][i]
        b = att_t['attribution_survey'][i]
        if a in channels and b in channels:
            channel_count[channels.index(a)] += 0.5
            channel_count[channels.index(b)] += 0.5
        elif a in channels and b not in channels:
            channel_count[channels.index(a)] += 1
        elif b in channels and a not in channels:
            channel_count[channels.index(b)] += 1
        else:
            continue
    result.loc[t] = channel_count
    
result
```

```python
result.to_csv(r'/Users/yankeyu/Desktop/channel_conversion.csv',  index = False, header=True)
```
